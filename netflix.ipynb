```python
data['Release Date']=pd.to_datetime(data['Release Date'])
data['Release Month']=data['Release Date'].dt.month
print(data['Release Month'])
```

```python
total_viewership=data.groupby('Release Month')['Hours Viewed'].sum().head(10)
print(total_viewership.index)
```

```python
sns.scatterplot(x=total_viewership.index)
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Total hours viewed in billions")
plt.title("Total Hours Viewership by Release Month")
plt.show()


```
